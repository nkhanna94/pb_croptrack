```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rain = pd.read_csv("data/AnnualAverageRainFall.csv")
rice = pd.read_csv("data/Production_Rice_1.csv")
wheat =  pd.read_csv("data/Production_Wheat_1.csv")
```

```python
datasets = {'Rainfall': rain, 'Rice': rice, 'Wheat': wheat}
```

```python
for name, df in datasets.items():
    print(f"\n===== {name} Dataset =====")
    print("Shape:", df.shape)
    print("Columns:", df.columns[:10].tolist(), "...")
    print(df.head(3))
    print(df.info())
    print("Missing values per column:\n", df.isna().sum().head(10))
```

```python
for name, df in datasets.items():
    numeric_cols = df.columns[1:]
    desc = df[numeric_cols].describe().T[['min','max','mean']]
    print(f"\n--- {name} Stats ---")
    print(desc.head(10))
```

```python
district = "Amritsar"

for name, df in datasets.items():
    df_t = df.set_index('District/Year').T
    if district in df_t.columns:
        plt.figure()
        plt.plot(df_t.index.astype(int), df_t[district])
        plt.title(f"{district} - {name} Trend")
        plt.xlabel("Year")
        plt.ylabel(name)
        plt.grid(True)
        plt.show()
```

* Wheat and rice follow a similar trend for amritsar.

```python
# Example for rainfall
rain_long = rain.melt(id_vars=['District/Year'], var_name='Year', value_name='Rainfall')
rain_long.rename(columns={'District/Year':'District'}, inplace=True)
rain_long['Year'] = rain_long['Year'].astype(int)

# Repeat for Rice and Wheat
rice_long = rice.melt(id_vars=['District/Year'], var_name='Year', value_name='Rice_Production')
rice_long.rename(columns={'District/Year':'District'}, inplace=True)
rice_long['Year'] = rice_long['Year'].astype(int)

wheat_long = wheat.melt(id_vars=['District/Year'], var_name='Year', value_name='Wheat_Production')
wheat_long.rename(columns={'District/Year':'District'}, inplace=True)
wheat_long['Year'] = wheat_long['Year'].astype(int)
```

```python
df = rain_long.merge(rice_long, on=['District','Year'], how='left') \
              .merge(wheat_long, on=['District','Year'], how='left')
```

```python
print(rice_long['Year'].min(), rice_long['Year'].max())
print(wheat_long['Year'].min(), wheat_long['Year'].max())
print(rain_long['Year'].min(), rain_long['Year'].max())
```

```python
common_years = range(1970, 2018)  
df = df[df['Year'].isin(common_years)]
```

```python
df.pivot_table(index='District', columns='Year', values='Rainfall').isna().sum(axis=1)
```

```python
# Fill Rainfall with district mean
df['Rainfall'] = df.groupby('District')['Rainfall'].transform(lambda x: x.fillna(x.mean()))

# Fill Rice/Wheat with district mean
df['Rice_Production'] = df.groupby('District')['Rice_Production'].transform(lambda x: x.fillna(x.mean()))
df['Wheat_Production'] = df.groupby('District')['Wheat_Production'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df_clean = df.copy()
df_clean = df_clean.dropna(subset=['Rainfall'])
```

```python
df_clean.isna().sum()
```

```python
df['District'].unique()
```

```python
df_clean = df_clean[df_clean['District'] != 'Punjab']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Check overall trends: Wheat and Rice Production vs Rainfall
plt.figure(figsize=(12,6))
sns.scatterplot(data=df_clean, x='Rainfall', y='Wheat_Production', hue='District', palette='tab20')
plt.title('Wheat Production vs Rainfall by District')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Wheat Production (kg/ha)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(12,6))
sns.scatterplot(data=df_clean, x='Rainfall', y='Rice_Production', hue='District', palette='tab20')
plt.title('Rice Production vs Rainfall by District')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Rice Production (kg/ha)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 2️⃣ Production trends over time
plt.figure(figsize=(14,7))
for district in df_clean['District'].unique():
    plt.plot(df_clean[df_clean['District']==district]['Year'], 
             df_clean[df_clean['District']==district]['Wheat_Production'], label=district)
plt.title('Wheat Production Trends by District')
plt.xlabel('Year')
plt.ylabel('Wheat Production')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()

plt.figure(figsize=(14,7))
for district in df_clean['District'].unique():
    plt.plot(df_clean[df_clean['District']==district]['Year'], 
             df_clean[df_clean['District']==district]['Rice_Production'], label=district)
plt.title('Rice Production Trends by District')
plt.xlabel('Year')
plt.ylabel('Rice Production')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()

# 3️⃣ Correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(df_clean[['Rainfall','Rice_Production','Wheat_Production']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Rainfall, Rice, and Wheat Production')
plt.show()
```

* Rice vs Wheat: very high correlation (0.977). → districts with high rice production also have high wheat production, likely reflecting fertile districts or overall cropped area.
